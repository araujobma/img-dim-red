# Image Dimensionality Reduction

Reducing the dimension of images by transforming images to grayscale(one channel of 0 to 255) and binary(0 or 255) images.

```python
import numpy as np
```

## Reading the original image

```python
# importing matplotlib modules
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
 

img = mpimg.imread('lena.jpeg')
 

plt.imshow(img)
```

Image has shape of 255*255 and 3 channels 

```python
img.shape
```

```python
#The first channel(red)
img[:,:,0]
```

## Plotting the three channels separately
We can accomplish this by making the other two channels have zero values for all pixels

```python
img_red = np.zeros(img.shape, dtype='uint8')
img_red[:,:,0] = img[:,:,0]
plt.imshow(img_red)
```

```python
img_green = np.zeros(img.shape, dtype='uint8')
img_green[:,:,1] = img[:,:,1]
plt.imshow(img_green)
```

```python
img_blue = np.zeros(img.shape, dtype='uint8')
img_blue[:,:,2] = img[:,:,2]
plt.imshow(img_blue)
```

## Getting grayscale image by averaging the three channels

```python
img_gray = np.average(img, axis=2)
img_gray
```

```python
#We need to define cmap as gray so the image is not displayed as single red or green or blue channel
plt.imshow(img_gray, cmap='gray')
```

## Creating a binary image

```python
img_gray.max()
```

```python
img_binary = (img_gray >= img_gray.mean()) * 255
```

```python
plt.imshow(img_binary, cmap='gray')
```
